```python
import Bio
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
from Bio import Align
from Bio import SeqIO
from Bio.Align import substitution_matrices

import pandas as pd
```

```python
##Make a function to define sequences from a fasta file.

def define_seq(pathway, form, reference_name, new_name):
    for seq in SeqIO.parse(pathway, form):
        if seq.name == reference_name:
            new_name = seq.seq
            return new_name
```

```python
##Define S.cerevisiae ERG11 sequence and C.Albicans as well

ScERG11 = define_seq(r"c:S288C_YHR007C_ERG11_protein.fsa","fasta","ERG11", "ScERG11" )
print(ScERG11)

CaERG11 = define_seq(r"c:CaERG11.fsa","fasta","ERG11", "CaERG11" )
print(CaERG11)
```

```python
##Prepare the alignments parameters and make the alignment

aligner = Align.PairwiseAligner()
aligner.open_gap_score = -10
aligner.extend_gap_score = -0.5
aligner.substitution_matrix = substitution_matrices.load("BLOSUM62")

alignments = aligner.align(ScERG11, CaERG11)
print(alignments[0])
```

```python
##Alignment function

def align_seq(target,query):
    aligner = Align.PairwiseAligner()
    aligner.open_gap_score = -10
    aligner.extend_gap_score = -0.5
    aligner.substitution_matrix = substitution_matrices.load("BLOSUM62")
    score = aligner.score(target, query)
    print(score)
    alignments = aligner.align(target,query)
    print(alignments[0]) 
    return alignments[0]

Sc_Ca_Align = align_seq(ScERG11,CaERG11)
```

```python
##Import a position list from the query sequence

pos_file = pd.read_excel(r"C:query_position.xlsx",sheet_name=0, header=None)
pos_file

```

```python
pos_list = list(pos_file[0]) 
```

```python
##From that, we can make a function that takes a list of query positions and 
#return it for the target.

def compare_pos_list(alignment, pos_list):
    
    targetposlist = []
    queryposlist = []
    targetAAlist = []
    queryAAlist = []
    
    for position in pos_list:
        position = position -1
        new_align = str(alignment)
        new_align_split = new_align.split("\n")
        query_align = new_align_split[0]
        target_align = new_align_split[2]
        
        target_tiret_count = target_align[0:position + 1].count("-")

        query_tiret_count = query_align[0:position + 1].count("-")
        query_aa_position = position + query_tiret_count  
 
        target_position = position - target_tiret_count + query_tiret_count
        
        targetposlist.append(target_position+1)
        queryposlist.append(position +1)
        targetAAlist.append(target_align[query_aa_position])
        queryAAlist.append(query_align[query_aa_position])
    
    dictionary = {'query_pos': queryposlist, 'query_aa': queryAAlist, 'target_pos': targetposlist, 'target_aa': targetAAlist}  
    dataframe = pd.DataFrame(dictionary) 
    dataframe.to_csv('compare_pos_list.csv')
    
    return dataframe
        
```

```python
Ca_SC_analysis = compare_pos_list(Sc_Ca_Align, pos_list)
Ca_SC_analysis.to_csv('pos_list.csv')
```

```python
####Make a big function that takes all of that (except the sequences)
##align+compare

def find_aligned_positions(query, target, pos_list):
        
    aligner = Align.PairwiseAligner()
    aligner.open_gap_score = -10
    aligner.extend_gap_score = -0.5
    aligner.substitution_matrix = substitution_matrices.load("BLOSUM62")

    alignments = aligner.align(query, target)
    alignment = alignments[0]
    
    targetposlist = []
    queryposlist = []
    targetAAlist = []
    queryAAlist = []
    
    for position in pos_list:
        position = position -1
        new_align = str(alignment)
        new_align_split = new_align.split("\n")
        query_align = new_align_split[0]
        target_align = new_align_split[2]
        
        target_tiret_count = target_align[0:position + 1].count("-")

        query_tiret_count = query_align[0:position + 1].count("-")
        query_aa_position = position + query_tiret_count  
 
        target_position = position - target_tiret_count + query_tiret_count
        
        targetposlist.append(target_position+1)
        queryposlist.append(position +1)
        targetAAlist.append(target_align[query_aa_position])
        queryAAlist.append(query_align[query_aa_position])
    
    dictionary = {'query_pos': queryposlist, 'query_aa': queryAAlist, 'target_pos': targetposlist, 'target_aa': targetAAlist}  
    dataframe = pd.DataFrame(dictionary) 
    dataframe.to_csv('compare_pos_list.csv')
    
    return dataframe
        
```
